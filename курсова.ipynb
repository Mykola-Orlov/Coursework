## Завдання 4  
Відобразити 10 елементів для 4-ох ознак Виділити
зеленим кольором максимальні елементи в таблиці, а червоним мінімальні для цих 10-ти
елементів.

```python
# Зчитування значень з файлу CSV
data = pd.read_csv('C:\\Users\\орлов ілля\\OneDrive\\Робочий стіл\\Anaconda\\курсач\\dataset\\Cancer_Data.csv')

# Вибір стовпців 3, 4, 5, 6 та перших 10 рядків
selected_data = data.iloc[:10, [2, 3, 4, 5]]

# Ініціалізація списків для збереження мінімальних та максимальних значень
min_vals_34 = selected_data.iloc[:, :2].min(axis=1)
max_vals_34 = selected_data.iloc[:, :2].max(axis=1)
min_vals_56 = selected_data.iloc[:, 2:].min(axis=1)
max_vals_56 = selected_data.iloc[:, 2:].max(axis=1)

# Створення карти для відображення даних
heatmap = np.zeros((10, 4, 4), dtype=float)

for i in range(10):
    for j in range(4):
        val = selected_data.iat[i, j]
        
        if j < 2:
            min_val = min_vals_34[i]
            max_val = max_vals_34[i]
        else:
            min_val = min_vals_56[i]
            max_val = max_vals_56[i]
        
        if val == min_val:
            heatmap[i, j] = [0, 1, 0, 1.0]  # Зелений колір
        elif val == max_val:
            heatmap[i, j] = [1, 0, 0, 1.0]  # Червоний колір
        else:
            heatmap[i, j] = [1, 1, 1, 0.0]  # Прозорий колір

# Встановлення більшого розміру фігури
plt.figure(figsize=(5, 15))

# Відображення даних у вигляді теплової карти
plt.imshow(heatmap)

# Додавання значень до комірок та збільшення розміру шрифту
for i in range(10):
    for j in range(4):
        val = selected_data.iat[i, j]
        plt.text(j, i, f'{val:.2f}', ha='center', va='center', color='black', fontsize=12)

# Вимкнення вісей та відображення графіка
plt.axis('off')
plt.show()
```

```python
# Визначаємо дані, які потрібно стандартизувати 
data = np.array([1, 2, 3, 4, 5])

# Обчислюємо математичне сподівання (середнє значення)
mean = np.mean(data)

# Обчислюємо стандартне відхилення
std = np.std(data)

# Стандартизуйемо дані
standardized_data = (data - mean) / std

# Виводимо результати
print("Математичне сподівання:", mean)
print("Стандартне відхилення:", std)
print("Стандартизовані дані:", standardized_data)
```

```python
# Визначаємо пару ознак з найбільшою кореляцією (окрім 1 та 2 стовпців)
max_correlation = 0
feature_A, feature_B = None, None

for i in range(2, len(df.columns)):                # Цикл, який перебирає індекси стовпців у фреймі даних, починаючи з індексу 2.
    for j in range(i + 1, len(df.columns)):      # Внутрішній цикл, який перебирає індекси стовпців у фреймі даних, починаючи з індексу, який є на одиницю більшим за i
        correlation = df[df.columns[i]].corr(df[df.columns[j]])  # Обчислення коефіцієнта кореляції між парою стовпців
        if abs(correlation) > max_correlation:     # Перевірка, чи модуль коефіцієнта кореляції є більшим за поточне максимальне значення.
            max_correlation = abs(correlation)     # Якщо умова вище виконується, оновлюємо max_correlation новим значенням
            feature_A, feature_B = df.columns[i], df.columns[j]  # Оновлення feature_A та feature_B іменами стовпців, які утворюють пару з максимальним або мінімальним коефіцієнтом кореляці

print(f"Найбільша кореляція (окрім 1 та 2 стовпців) між {feature_A} та {feature_B}: {max_correlation}")


# Збудовуємо relplot для цих двох ознак
sns.relplot(x=feature_A, y=feature_B, data=df)
plt.show()
```

## Завдання 9

Видалити також випадкові значення з ознак що мають найбільшу кореляцію у кількості 10%
для обох. Підрахувати кореляцію. Збудувати multipple scatterplot за допомогою FacetGrid
для цих оновлених ознак використовуючи бібліотеку seaborn взявши третю ознаку за вісь x.

```python
# Визначення пари ознак з найбільшою кореляцією (окрім 1 та 2 стовпців)
max_correlation = 0
feature_A, feature_B = None, None

for i in range(2, len(df.columns)):                # Цикл, який перебирає індекси стовпців у фреймі даних, починаючи з індексу 2.
    for j in range(i + 1, len(df.columns)):      # Внутрішній цикл, який перебирає індекси стовпців у фреймі даних, починаючи з індексу, який є на одиницю більшим за i
        correlation = df[df.columns[i]].corr(df[df.columns[j]])  # Обчислення коефіцієнта кореляції між парою стовпців
        if abs(correlation) > max_correlation:     # Перевірка, чи модуль коефіцієнта кореляції є більшим за поточне максимальне значення.
            max_correlation = abs(correlation)     # Якщо умова вище виконується, оновлюємо max_correlation новим значенням
            feature_A, feature_B = df.columns[i], df.columns[j]  # Оновлення feature_A та feature_B іменами стовпців, які утворюють пару з максимальним або мінімальним коефіцієнтом кореляці

print(f"Найбільша кореляція (окрім 1 та 2 стовпців) між {feature_A} та {feature_B}: {max_correlation}")

# Випадкове видалення 10% значень для обох ознак з найбільшою кореляцією
df_no_missing = df.copy()

for feature in [feature_A, feature_B]:   # Це цикл, який перебирає стовпці, які були знайдені з найбільшим або найменшим коефіцієнтом
    indices_to_remove = np.random.choice(df_no_missing.index, size=int(0.1 * len(df_no_missing)), replace=False) # вказує розмір (кількість) індексів, які потрібно вибрати, і в даному випадку це 10% від усіх рядків.
    df_no_missing.loc[indices_to_remove, feature] = np.nan   # Встановлення випадково обраних рядків та стовпця feature у фреймі даних 

# Підрахунок кореляції після видалення випадкових значень
correlation_after_removal = df_no_missing[feature_A].corr(df_no_missing[feature_B])
print(f"Кореляція після видалення випадкових значень: {correlation_after_removal}")

# Збудова multipple scatterplot за допомогою FacetGrid
g = sns.FacetGrid(df_no_missing, col="diagnosis", height=4, aspect=1) # aspect=1 - на співвідношення ширини та висоти.
g.map(sns.scatterplot, feature_A, feature_B, alpha=0.7) # alpha=0.7 вказує на прозорість точок
g.add_legend()

plt.show()
```
